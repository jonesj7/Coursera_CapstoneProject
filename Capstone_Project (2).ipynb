More than one neighborhood can exist in one postal code area. For example, in the table on the Wikipedia page, you will notice that M5A is listed twice and has two neighborhoods: Harbourfront and Regent Park. These two rows will be combined into one row with the neighborhoods separated with a comma.

```python
num=0
row = []
past_code = ''
amend_loc = 0

for n in range(df_canada.shape[0]):
    for item in df_canada.iloc[n,0:3]:
        current_code = df_canada.iloc[n,0]
        
        if(current_code == past_code):
            if(num == 2):                
                df_canada.iloc[amend_loc,2] = (df_canada.iloc[amend_loc,2]+ ", " + df_canada.iloc[n,2])
                df_canada.iloc[n,2] = np.nan 
                num = 0 
            else:
                num = num+1
            row.append(item)
        else:           
            row.append(item)
            num = num+1
            amend_loc = n
    past_code = current_code
    row=[]
    num = 0

df_canada.dropna(axis=0, inplace=True)
df_canada.reset_index(drop=True, inplace=True)
df_canada.head()
```

```python
LIMIT = 100 # limit of number of venues returned by Foursquare API
radius = 500 # define radius

def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighbourhood', 
                  'Neighbourhood Latitude', 
                  'Neighbourhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
num_top_venues = 5

for hood in Toronto_grouped['Neighbourhood']:
    print("----"+hood+"----")
    temp = Toronto_grouped[Toronto_grouped['Neighbourhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighbourhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighbourhood'] = Toronto_grouped['Neighbourhood']

for ind in np.arange(Toronto_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(Toronto_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

```python
# set number of clusters
kclusters = 5

Toronto_grouped_clustering = Toronto_grouped.drop('Neighbourhood', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(Toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
# add clustering labels
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

Toronto_merged = Toronto_data

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
Toronto_merged = Toronto_merged.join(neighborhoods_venues_sorted.set_index('Neighbourhood'), on='Neighbourhood')

Toronto_merged.head() # check the last columns!
```

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(Toronto_merged['Latitude'], Toronto_merged['Longitude'], Toronto_merged['Neighbourhood'], Toronto_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```

## Introduction
Self-determination was a major point of distinction contained in the 1960 U.N. Declaration on the Granting of Independence to Colonial Countries and Peoples. As that decolonial wave swept the world some countries such as India immediately gained self determination in 1947 while a country such as Zimbabwe didn't immediately gain it's self determination until after additional struggle against European settlers who formed the local governing powers of Rhodesia. That distinction between self determination from foreign European colonial powers V.S. self determination from local European settler governance forms the context for the investigation here which takes the form of the following question. When will indigenous peoples of the Americas regain self-determination and political control over their respective lands/nations?

This Question takes on special significance in relation to the African Diaspora in the Americas. When the indigenous peoples regain political self-determination, that African population has to position themselves for a new political reality. When looking at the Post WW2 landscape a wave of countries gained political freedom via the decolonial process. That said the local indigenous people were not always the wielders of that political power which caused additional conflict that can be avoided if predicted. 

Examples of such conflicts being the various conflicts in Fiji between indigenous Fijians and Fijians of Indian origin. The aforementioned Zimbabwe/Rhodesia where the 1964-1979 "bush war" occurred to expel European settlers in control of the government. Ugandan expulsion of Asians in 1972 who were in control of the economy. Namibian War of Independence between SWAPO and the European settlers who made up Apartheid South Africa from 1966-1990. The common thread here being the settler population that exists after the external colonial powers have move on from direct political oversight to take up a neocolonial stance. If not careful the African diaspora of the Americas stands to be that settler population hindering the indigenous people of the Americas from taking rightful control of their own political and economic reality. 

This preliminary study serves as a proof of concept for establishing a basic framework for development of a "Time Frame Till Indigenous Governance" that can later be build upon for greater accuracy.

## Data

Training and testing data used in this proof of concept primarily comes from a scrape of the wikipedia decolonization page. 
Preliminary data utilized here will be Year, Colonizer, Decolonized_State, Occupied, Colonized, Settled, Enslaved, Forced_Trade, and Region.
* Year - year of Indigenous Governance 
* Colonizer - external state governing over the local populous
* Decolonized_State - country that gained Indigenous Governance
* Occupied - was the country overrun by foreign troops (EX: China)
* Colonized - was the country governed remotely by an external state (EX: India)
* Settled - was the country governed locally by foreign citizens (EX: South Africa)
* Enslaved - was the local populous enslaved by settlers/colonizer (EX: Ghana)
* Forced_Trade - was the country forced to trade by an external state (EX: Japan)
* Region - Africa, EastAsia, India, SouthEastAsia, Oceania, Europe, NorthAmerica, Caribbean, CentralAmerica, SouthAmerica, CentralAsia  

Mapping data of the countries involved will come from foursquare API. Specifically longitude and latitude.
